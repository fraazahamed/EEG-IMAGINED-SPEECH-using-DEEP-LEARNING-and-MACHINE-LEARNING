```python
def set_label(epochss):
    labels = epochss.events[:, -1].copy()
    event_id = epochss.event_id
    for i in range(len(labels)):
        label_key =list(event_id.keys())[list(event_id.values()).index(labels[i])]
        start_index =label_key.find("-")
        label_key=label_key[start_index+1:]
        labels[i] =imagined_words_dict[label_key]
    return labels
```

```python
# labels1 = set_label(epochs1)
# labels2 = set_label(epochs2)
# labels = np.array(labels1.tolist() + labels2.tolist())
labels = set_label(epochs)
```

```python
def make_windowing(X,Y):
    X_new = []
    Y_new = []
    npt = 128 # .5 sec
    stride = 8 # around .1 sec
    ctr = 0
    for i in range(0,X.shape[0]):
        y = Y[i]
        a= X[i,:,:]
        a = a.transpose()
        a.shape
        val = 0
        kd=len(a)
        while val<=(len(a)-npt):
            x = a[val:val+npt,:]

            X_new.append(x.T)
            Y_new.append(y)
            val = val+stride
            
    return np.array(X_new),np.array(Y_new)
```

```python
window_data,window_labels = make_windowing(data,labels)
#window_data,window_labels =data,labels
```

```python
from scipy import stats
def mean(data):
    return np.mean(data,axis=-1)
    
def std(data):
    return np.std(data,axis=-1)

def ptp(data):
    return np.ptp(data,axis=-1)

def var(data):
        return np.var(data,axis=-1)

def minim(data):
      return np.min(data,axis=-1)


def maxim(data):
      return np.max(data,axis=-1)

def argminim(data):
      return np.argmin(data,axis=-1)


def argmaxim(data):
      return np.argmax(data,axis=-1)

def mean_square(data):
      return np.mean(data**2,axis=-1)

def rms(data): #root mean square
      return  np.sqrt(np.mean(data**2,axis=-1))  

def abs_diffs_signal(data):
    return np.sum(np.abs(np.diff(data,axis=-1)),axis=-1)


def skewness(data):
    return stats.skew(data,axis=-1)

def kurtosis(data):
    return stats.kurtosis(data,axis=-1)

def concatenate_features(data):
    return np.concatenate((mean(data),std(data),ptp(data),var(data),minim(data),maxim(data),argminim(data),argmaxim(data),
                          mean_square(data),rms(data),abs_diffs_signal(data),
                          skewness(data),kurtosis(data)),axis=-1)
```

```python
from tqdm import tqdm
features=[]
for datta in tqdm(window_data):
    features.append(concatenate_features(datta))
features=np.array(features)
features.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# Assuming you have your feature matrix X and target labels y

# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, window_labels, test_size=0.25, random_state=50)

```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Create a KNN classifier with k=5
k = 5
clf = KNeighborsClassifier(n_neighbors=k)
```

```python

# Train the classifier on the training data
clf.fit(train_features, train_labels)
```

```python
# Make predictions on the test data
y_pred = clf.predict(test_features)
```

```python
# Calculate the accuracy of the classifier
accuracy = accuracy_score(test_labels, y_pred)
print("Accuracy:", accuracy)
```
